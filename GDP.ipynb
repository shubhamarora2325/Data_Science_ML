```python
#Calculate average growth of states for the  duration 2013-14, 2014-15 and 2015-16
years = ['2013-14','2014-15','2015-16']
df_per = df_GDP.loc[(df_GDP['Items  Description'] == '(% Growth over previous year)')]
```

```python
df_per.head()
```

```python
data_avg = df_per.loc[df_per.Duration.isin(years)]
```

```python
avg = round(data_avg.mean(),2).sort_values(ascending=True)
```

```python
print(avg)
```

```python
clrs = []
for i in avg.values:
    if i==avg['All_India GDP']:
        clrs.append('b')
    elif i > avg['All_India GDP']:
        clrs.append('g')
    else:
        clrs.append('r')
fig = plt.figure(figsize = (8,8))
graph = fig.add_subplot(1,1,1)
graph.set(title = 'Average growth of states for the duration 2013-14,2014-15 and 2015-16',xlabel = 'mean percentage',ylabel = 'States')
graph.barh(avg.index,avg.values,color = clrs)
plt.grid()
plt.show()
```

```python
#PART - 1-B(2)
# Concatenate the csv files and adding it to dataframe
#Filter out the union territories (Delhi, Chandigarh, Andaman and Nicobar Islands, etc.) for further analysis, 
#as they are governed directly by the centre, not state governments.
```

```python
states = ['Andhra_Pradesh', 'Arunachal_Pradesh', 'Assam', 'Bihar',
       'Chhattisgarh', 'Goa', 'Gujarat', 'Haryana', 'Himachal_Pradesh',
       'Jammu & Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Madhya_Pradesh',
       'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha',
       'Punjab', 'Rajasthan', 'Sikkim', 'Tamil_Nadu', 'Telangana', 'Tripura',
       'Uttar_Pradesh', 'Uttarakhand', 'West_Bengal']


os.chdir("G:/GDP Assignment Data/States")
filelist=glob.glob("*.csv")
dfList = []
    
for filename in filelist:
    for nameofstate in states:
        if nameofstate in filename:
            file_to_read = filename
            print(file_to_read)
            df=pd.read_csv(file_to_read,encoding ='latin1')
            #print(dfList)
            df= df[['Item', '2014-15']]
            df = df.transpose()
            #print(df)
            header=df.iloc[0]
            df=df[1:]
            df.columns=header
            #index_name_list = filename.split("-")
            #print(index_name_list)
            #index_name = index_name_list[1]
            #print(index_name)
            if nameofstate=='Uttar_Pradesh':
                df['state'] = 'Uttar Pradesh'
            elif nameofstate=='Madhya_Pradesh':
                df['state'] = 'Madhya Pradesh'
            elif nameofstate=='Andhra_Pradesh':
                df['state'] = 'Andhra Pradesh'   
            elif nameofstate=='Arunachal_Pradesh':
                df['state'] = 'Arunachal Pradesh'   
            elif nameofstate=='Tamil_Nadu':
                df['state'] = 'Tamil Nadu'
            elif nameofstate == 'Himachal_Pradesh':
                df['state'] = 'Himachal Pradesh'
            else:      
                df['state'] = nameofstate
                
            dfList.append(df)
                
            concatdf = pd.concat(dfList,axis = 0)
    
    
```

```python
# PART -1 -B (4)
#Plot the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states.
```

```python
def percent_gdp(sector, total):
    return sector/total
    
concatdf['primary_gdp'] = concatdf[['Primary','Gross State Domestic Product']].apply(lambda x: percent_gdp(*x), axis=1)
concatdf['secondary_gdp'] = concatdf[['Secondary','Gross State Domestic Product']].apply(lambda x: percent_gdp(*x), axis=1)
concatdf['tertiary_gdp'] = concatdf[['Tertiary','Gross State Domestic Product']].apply(lambda x: percent_gdp(*x), axis=1)
```
